Soma de listas

```python

# Define uma lista de números inteiros
numeros = [1,2,3]
# Inicializa a variável 'soma' com zero
soma = 0
# Inicializa a variável 'i' (contador) com zero
i = 0
# Loop 'while' para iterar sobre a lista 'numeros'
# O loop continua enquanto 'i' for menor que o tamanho da lista
while i < len(numeros):
  # Adiciona o número atual da lista à variável 'soma'
  soma += numeros[i]
  # Incrementa o contador 'i' para passar para o próximo número
  i += 1
# Imprime o valor final da soma
print(soma)
```

Eh primo

```python
import math

# Solicita ao usuário que digite um número e o converte para inteiro
n = int(input('Digite um número: '))

def eh_primo(n):
  # Verifica se um número é primo.
  # Um número é primo se for maior que 1 e não tiver divisores além de 1 e ele mesmo.
  # Esta função testa a divisibilidade apenas até a raiz quadrada do número para otimizar a performance.
  if n <= 1:
    return False  # Números menores ou iguais a 1 não são primos

  # Itera de 2 até a raiz quadrada de 'n' (inclusive).
  # Se 'n' for divisível por qualquer número nesse intervalo, então não é primo.
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False  # Encontrou um divisor, então 'n' não é primo

  return True  # Se nenhum divisor foi encontrado, 'n' é primo

# Chama a função eh_primo com o número inserido pelo usuário e imprime o resultado
if eh_primo(n):
  print(f'{n} é um número primo.')
else:
  print(f'{n} não é um número primo.')

Digite um número: 5
5 é um número primo.
```

Inversão de listas

```python

# Define a função para inverter uma string (texto) usando um loop
def inverte_texto_com_loop(texto):
  invertido = ''  # Inicializa uma string vazia para armazenar o texto invertido
  for letra in texto:  # Itera sobre cada letra do texto de entrada
    invertido = letra + invertido  # Adiciona a letra atual no início da string 'invertido'
  return invertido  # Retorna o texto invertido

# Imprime o texto invertido, chamando a função com a string "Pedro"
print(f'Nome invertido (com loop): {inverte_texto_com_loop("Gabriella")}')

# Define a função para inverter uma string (texto) de forma mais eficiente usando fatiamento de string
def inverte_texto_com_fatiamento(texto):
  return texto[::-1]  # Retorna a string invertida usando fatiamento (slice) de string

# Imprime o texto invertido, chamando a função com a string "Pedro" para a versão com fatiamento
print(f'Nome invertido (com fatiamento): {inverte_texto_com_fatiamento("Gabriella")}')

```
